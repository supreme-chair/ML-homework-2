In this dataset we stumbled upon problem with Apple devices, which ran on Windows, and devices from other companies, which ran on Mac OS. This can be explained by custom computers, but we ultimatly deicded to delete rows, containing this strange data

```python
import pandas as pd

df = pd.read_csv('computer_prices_all.csv')

print("SIZE BEFORE REMOVAL:", df.shape)
print("Number of rows before:", len(df))

df_clean = df[
    ~(
        ((df['brand'] == 'Apple') & (df['os'] != 'macOS')) |  # Apple without macOS
        ((df['brand'] != 'Apple') & (df['os'] == 'macOS'))    # Non-Apple with macOS
    )
]

print("\nSIZE AFTER REMOVAL:", df_clean.shape)
print("Number of rows after:", len(df_clean))
print("Rows removed:", len(df) - len(df_clean))

print("\nVERIFICATION:")
print("1. Apple devices (should only have macOS):")
apple_check = df_clean[df_clean['brand'] == 'Apple']['os'].unique()
print(f"   OS for Apple: {apple_check}")

print("\n2. Devices with macOS (should only be Apple):")
macos_check = df_clean[df_clean['os'] == 'macOS']['brand'].unique()
print(f"   Brands with macOS: {macos_check}")

df_clean.to_csv('computer_prices_clean.csv', index=False)
print(f"\nCleaned dataset saved to: computer_prices_clean.csv")
```
